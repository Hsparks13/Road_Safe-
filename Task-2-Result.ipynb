```python
X_train = np.zeros((len(train_files),256,256))
```

```python
Y_train=list()
```

```python
i=0
for fi in train_files:
    im=cv2.imread(fi, cv2.IMREAD_GRAYSCALE)
    new_array=cv2.resize(im,(256,256))
    X_train[i,:,:]=new_array
    label=(train_files[i].partition(key1)[2]).partition(key2)[0]
    Y_train[i]= label
    i=i+1
```
